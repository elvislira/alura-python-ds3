## Aula 02 - Exploração dos dados

* As datas|

```python
datas = dados_transpostos[:,0]
datas
```

* Todos os preços

```python
precos = dados_transpostos[:,1:6]
precos
```

Gerando um gráfico:

```python
import matplotlib.pyplot as plt

plt.plot(datas, precos[:, 0])
```

* O que acontece é que essas datas estão sendo interpretadas como números. Ou seja, janeiro de 2013 (1.2013) está sendo lindo como 1,2013.

    Vamos corrigir o problema:

```python
datas = np.arange(1, 88)

plt.plot(datas, precos[:,0])
```

* Dados das cidades

```python
moscow = precos[:, 0]
kaliningrad = precos[:, 1]
petersburg = precos[:, 2]
krasnodar = precos[:, 3]
ekaterinburg = precos[:, 4]

plt.plot(datas, moscow)
plt.plot(datas, kaliningrad)
plt.plot(datas, petersburg)
plt.plot(datas, krasnodar)
plt.plot(datas, ekaterinburg)

plt.legend(['Moscow', 'Kaliningrad', 'Petersburg', 'Krasnodar', 'Ekaterinburg'])
```

* Analisando Moscow ano a ano:

```python
meses = np.arange(1, 13)
anos = np.arange(2013, 2020)

ano2013 = moscow[0:12]
ano2014 = moscow[12:24]
ano2015 = moscow[24:36]
ano2016 = moscow[36:48]
ano2017 = moscow[48:60]
ano2018 = moscow[60:72]
ano2019 = moscow[72:84]

plt.plot(meses, ano2013)
plt.plot(meses, ano2014)
plt.plot(meses, ano2015)
plt.plot(meses, ano2016)
plt.plot(meses, ano2017)
plt.plot(meses, ano2018)
plt.plot(meses, ano2019)

plt.legend(anos)
```

```python
if np.allclose(ano2017, ano2019, 0.01):
    print('Pouca variação entre 2017 e 2019')
else:
    print('Variação em mais de 1%')
```
